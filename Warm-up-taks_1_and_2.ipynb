### Task 1

The first task consists of fitting the following logarithmic equation to a set of 50 (x,y) data points:

 y = a ⋅ log(x) + b

Your curve should be optimal with respect to the residual sum of squares metric (RSS). Perform the following steps


### -Load and plot the data

```python
data = pd.read_csv('p3_task-1.csv')

x = data['x'].values.reshape(-1,1) # (mental note) this converts an matrice into a column
y = data['y'].values

plt.scatter(x, y, label='Original Data')
plt.title('Original Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

### -Define a new variable log(x) and transform the data.

```python
x_log = np.log(x)
```

### -Fit a model to the new data


```python
model = LinearRegression()
model.fit(x_log, y)

a = model.coef_[0].round(2) #(mental note) obtain the coeficient 'a'. that would be the slope
b = model.intercept_.round(2) #(mental note) obtains the intercept, 'b' that ordered to origin

print(f"Ajusted model: y = {a} * log(x) + {b}")  
      
      
      
      
```

### -Plot the logarithmic curve against the original data (see plot above).


```python
# Generate values between 1 and 200 hundred
x_range = np.linspace(x.min(), x.max(), 200).reshape(-1,1)
# Compute the associate y values
y_prediction_curve = model.predict(np.log(x_range))

# Plot the line
plt.scatter(x, y)  # Plot the actual data
plt.plot(x_range, y_prediction_curve, c="C3", label="Logaritmic Adjustment")  # Plot the prediction line
plt.title('Logaritmic Adjustment from the Original Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

### -Compute the RSS and R2 measures.

```python
y_pred = model.predict(x_log)

RSS = round(np.sum((y - y_pred)**2), 2)  # Usar round() de Python, no .round() de NumPy
r2 = round(r2_score(y, y_pred), 2)       # Usar round() de Python, no .round() de NumPy

print("RSS", RSS)
print("R^2:", r2)
```

### -Discuss the results, is it a good R2 score?

```python
if r2 > 0.9:
    print("The Adjustment is really good")
elif r2 > 0.7:
    print("The Adjustment is reasonable")
else:
    print("The Adjustment is not reeally well done")
```

The goal is to build 3 models and compare different approaches to handle outliers:

A linear regression to the data with outliers,
A huber regression to the data with outliers,
A linear regression to the data without outliers.
You should

Start by using the train/test split methodology (e.g. 80-20 splits) to create a test set and keep it aside.
Plot the training data to get a quick overview of the data.
Build and apply a cleaning strategy to remove outliers from the training data.
Finally compare the test MAE scores for your 3 models and comment.
Note that we don’t remove outliers from the test data. This ensures that we use the same test data each time and thus the evaluations of the models are consistent.

The linear regression without outliers and huber regression models should both perform better than the linear regression with outliers.

```python
data = pd.read_csv('p3_task-2.csv')

print(f"data set: {data.shape}")

print(f"colums: {list(data.columns)}")

data.columns [-1] # mental note ( -1) is for the last position, if I have 10 colums the last would -9

print(f"objective variable: {data.columns [-1]}")

print(f"data set: {data.head()}")

print(f"data set: {data.describe()}\n")


x = data.drop('y', axis = 1)
y = data['y']
    
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 100)
print("split data 80-20\n")
print(f"training data: {x_train.shape[0]}")
print(f"test data: {x_test.shape[0]}")
  
 

plt.figure(figsize=(10, 8))
plt.subplot(2, 2, 1)
x_train.boxplot()
plt.title("Exploration Distribution Data")
plt.ylabel("Values")
    
    
plt.subplot(2, 2, 2)
plt.scatter(x_train['x1'], y_train)
plt.title("X1 versus Y")
plt.xlabel("X1")
plt.ylabel("Y")
  
plt.subplot(2, 2, 3)
plt.scatter(x_train['x2'], y_train)
plt.title("X2 versus Y")
plt.xlabel("X2")
plt.ylabel("Y")
  

plt.subplot(2, 2, 4)
plt.scatter(x_train['x3'], y_train)
plt.title("X3 versus Y")
plt.xlabel("X3")
plt.ylabel("Y")
plt.tight_layout()
       
plt.show()  
    
for feature in['x1', 'x2', 'x3']:
        Q1 = x_train[feature].quantile(0.25)
        Q3 = x_train[feature].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = x_train[(x_train[feature] < lower_bound) | (x_train[feature] > lower_bound)]
        print(f"Outliers found: {feature}: {len(outliers)}")

original_size = len(x_train)
        
x_train_cleaned = x_train.copy()
y_train_cleaned = y_train.copy()

for feature in['x1', 'x2', 'x3']:
        Q1 = x_train_cleaned[feature].quantile(0.25)
        Q3 = x_train_cleaned[feature].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        filter = (x_train_cleaned[feature] >= lower_bound) & (x_train_cleaned[feature] <= upper_bound)
        
        x_train_cleaned = x_train_cleaned[filter]
        y_train_cleaned = y_train_cleaned[filter] 

removed_data = original_size - len(x_train_cleaned)

print(f"Original Size: {original_size}")
print(f"Data after cleaning: {len(x_train_cleaned)}")    
print(f"Removed Outliers: {removed_data}")
```

```python
# Entrenar modelo Huber con datos originales (con outliers)
huber_with_outliers = HuberRegressor(epsilon=1.35, max_iter=100)
huber_with_outliers.fit(x_train, y_train)

# Predicciones
y_pred_huber_with_outliers = huber_with_outliers.predict(x_test)

# Métricas
mae_huber_with_outliers = mean_absolute_error(y_test, y_pred_huber_with_outliers)
r2_huber_with_outliers = r2_score(y_test, y_pred_huber_with_outliers)

print(f"Huber Regression CON outliers:")
print(f"  - MAE: {mae_huber_with_outliers:.2f}")
print(f"  - R²: {r2_huber_with_outliers:.4f}")
```

```python

# Entrenar modelo con datos limpios (sin outliers)
linear_without_outliers = LinearRegression()
linear_without_outliers.fit(x_train_cleaned, y_train_cleaned)

# Predicciones
y_pred_linear_without_outliers = linear_without_outliers.predict(x_test)

# Métricas
mae_linear_without_outliers = mean_absolute_error(y_test, y_pred_linear_without_outliers)
r2_linear_without_outliers = r2_score(y_test, y_pred_linear_without_outliers)

print(f"Linear Regression SIN outliers:")
print(f"  - MAE: {mae_linear_without_outliers:.2f}")
print(f"  - R²: {r2_linear_without_outliers:.4f}")
```

```python
print(f"\n🔍 EVALUACIÓN DE MAE SCORES:")
print(f"  - Linear con outliers MAE: {mae_linear_with_outliers:.2f}")
print(f"  - Huber con outliers MAE: {mae_huber_with_outliers:.2f}")
print(f"  - Linear sin outliers MAE: {mae_linear_without_outliers:.2f}")

# Verificar si los últimos dos modelos son mejores que el primero
if mae_huber_with_outliers < mae_linear_with_outliers:
    print(f"  ✅ Huber es mejor que Linear con outliers")
else:
    print(f"  ❌ Huber NO es mejor que Linear con outliers")

if mae_linear_without_outliers < mae_linear_with_outliers:
    print(f"  ✅ Linear sin outliers es mejor que Linear con outliers")
else:
    print(f"  ❌ Linear sin outliers NO es mejor que Linear con outliers")
```
